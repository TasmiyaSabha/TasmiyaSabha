```python
import pandas as pd
import numpy as np
import datetime
```

**Creating a table**

```python
py_students=['Ariba','Pia','Siya','Zuhan']
```

```python
py_scores=[67,98,45,78]
```

```python
py_er=pd.DataFrame(
    {
        'Name':py_students,
        'Marks':py_scores
    }
)
py_er
```

**Adding Columns**

```python
Gender=['F','F','F','M']
py_er['Gender']=Gender
py_er
```

**Removing columns**

```python
py_er.drop('Marks',axis=1)
```

**Renaming**

**Dealing with Missing values**

```python
miss_v=pd.DataFrame(
    {
        'x':[5,6,7],
     'y':[np.nan,np.nan,8],
     'z':[9,10,np.nan]
    }
)
miss_v
```

```python
miss_v.isna()
```

```python
miss_v.fillna(0)
```

```python
miss_v.dropna(axis=1)
```

**Using group by**

```python
py_er.groupby('Gender').mean()
```

```python
mylist=[1,8,3,2]
py_er.groupby(mylist).get_group(8)
```

```python
#Marks scored greater than 90
goodmarks=py_er['Marks']>50
goodmarks
```

```python
#Mean scores of students who scored above 80
py_er.groupby(goodmarks).mean()
```

**Merging and Joining**

```python
merg=pd.DataFrame(
    {
        'Name':['Ariba'],
        'ph.no':['+91 9876543222']
    }
)
merg
```

```python
py_er.merge(merg,left_on='Name',right_on='Name')
```

```python
py_er.merge(merg,left_on='Name',right_on='Name',how='outer')
```
