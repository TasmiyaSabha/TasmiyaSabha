```python
crimes = crime.resample('10AS').sum()

# Uses resample to get the max value only for the "Population" column
population = crime['Population'].resample('10AS').max()

# Updating the "Population" column
crimes['Population'] = population

crimes
```

```python
#2. Write a Python program to extract PRESENT year, month, date and time using Lambda.
import datetime
currentDateTime = datetime.datetime(2021,10,17)
print(currentDateTime)
year = lambda x: x.year
month = lambda x: x.month
day = lambda x: x.day
print('Year - ',year(currentDateTime))
print('Month - ',month(currentDateTime))
print('Day - ',day(currentDateTime))
 
```

```python
#3.Write a Python program that sum the length of the names of a given list of names after removing the names that starts with a lowercase letter. Use lambda function

py_names = ['Rihana','pihu',"Pooja",'FARU']
py_names=list(filter(lambda el:el[0].isupper() and el[1:].islower(),py_names))
print("Result:")
print(len(''.join(py_names)))
```

```python
#4. Write a Python program to extract specified size of strings from a give list of string values using lambda
def extract_string(str_list1, length):
    result = list(filter(lambda e: len(e) == length, str_list1))
    return result
str_list1 = ['Rihana','pihu',"Pooja",'Fahadh']
print("Original list:",str_list1)
length= 4
print("\nlength of the string to extract:",length)
print("\nAfter extracting strings of specified length from the said list:",extract_string(str_list1 , length))
```
